```python
import csv
tags = []
with open('tag_distinct.csv', 'r') as f:
    for line in f:
        tags.append(line.strip())
del tags[0]
```

```python
from datetime import datetime, timedelta
start = "2013-12-31"
end = "2017-06-15"
s = datetime.strptime(start, "%Y-%m-%d")
e = datetime.strptime(end, "%Y-%m-%d")
dates = []
while s <= e:
    sf = s.strftime("%Y-%m-%d")
    dates.append(sf)
    s += timedelta(1)
```

```python
import pandas as pd
df = pd.DataFrame([list([0.0001]*len(tags))]*len(dates),index=dates,columns=tags)
```

```python
with open('tag_date_rate.csv', 'r') as f:
    for line in csv.reader(f):
        try:
            df.set_value(line[1],line[0],float(line[2]))
        except ValueError:
            df.set_value(line[1],line[0],0)
df
```

```python
df.to_csv('tag_pandas.csv', sep=',', encoding='utf-8',index_label=False)
```
