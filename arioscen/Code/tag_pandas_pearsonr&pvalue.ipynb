```python
import csv
tags = []
with open('tag_distinct.csv', 'r') as f:
    for line in f:
        tags.append(line.strip())
del tags[0]
tags
```

```python
import pandas as pd
pm = pd.DataFrame([list([0.0001]*len(tags))]*len(tags),index=tags,columns=tags)
pv = pd.DataFrame([list([0.0001]*len(tags))]*len(tags),index=tags,columns=tags)
```

```python
import pandas as pd
df = pd.read_csv('tag_pandas_cut.csv')
df
```

```python
len(df.columns)
df.columns[0]
# df.iloc[:,1423].name
```

```python
from scipy.stats.stats import pearsonr

for i in range(0,len(df.columns)):
    i_name = df.columns[i]
    i_list = df.iloc[:,i]
    for j in range(0,len(df.columns)):
        j_name = df.columns[j]
        j_list = df.iloc[:,j]
        pear = pearsonr(i_list,j_list)
        pm_value = round(pear[0],4)
        pv_value = round(pear[1],4)
        pm.set_value(i_name,j_name,pm_value)
        pv.set_value(i_name,j_name,pv_value)
```

```python
pm.to_csv('tag_pandas_pearsonr.csv', sep=',', encoding='utf-8',index_label='tag')
```
