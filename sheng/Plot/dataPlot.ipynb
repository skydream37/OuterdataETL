```python
from pymongo import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei'] 
```

```python
client = MongoClient('localhost',27018)  
db = client['data']   
```

```python
shops = ["daiso","epice","gohappy","mft","poya","supergood"]
tags = ["雨傘" ,"雨衣" ,"收納盒" ,"貼紙" ,"化妝水" , "洗髮乳"]
```

```python
for shop in shops:
    count = []
    for tag in tags:
        collection=db[shop]
        a = collection.find({"Tags":tag}).count()
        count.append(a)
#======================draw time!========================
    plt.pie(count,labels=tags,colors=['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'],autopct='%1.1f%%', shadow=True, startangle=90)
    plt.axis('equal')    
    plt.show()
    print(shop,count)
```
